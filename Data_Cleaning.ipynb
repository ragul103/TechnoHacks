Peform Data Cleaning
Clean a Dataset by Removing Missing values and Outliers


```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Employee Sample Data.csv', encoding='latin1')
df
```

```python
df.info
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
def remove_outliers_zscore(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return data[(z_scores < threshold).all(axis=1)]
```

```python
numerical_columns = df_cleaned.select_dtypes(include=[np.number])
numerical_columns
```

```python
df_cleaned[numerical_columns.columns] = remove_outliers_zscore(numerical_columns)
df_cleaned[numerical_columns.columns] 
```

```python
df_cleaned.to_csv('cleaned_dataset.csv', index=False)
```
