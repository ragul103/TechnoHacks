Remove Duplication
Identify and remove duplicate values in a
dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df
```

```python
df.info()
```

```python
duplicate_rows = df[df.duplicated()]
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned
```

```python
df_cleaned.to_csv('Iris.csv', index=False)
```
